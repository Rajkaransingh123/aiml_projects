```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/cleaned_email_sms_data.csv')
```

```python
df
```

```python
df.drop('Unnamed: 0',axis =1, inplace =True)
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label'], test_size=0.2, random_state=42
)
```

### Converts raw text into vectors so the model can understand

```python
tfidf = TfidfVectorizer(max_features=3000)
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec  = tfidf.transform(X_test)
```

#### build and train model

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

### Prediction and evaluation

```python
y_pred = model.predict(X_test_vec)
train_pred = model.predict(X_train_vec)
```

```python
print(f"Train Accuracy: {accuracy_score(y_train,train_pred)}")
print(f"Test Accuracy: {accuracy_score(y_test,y_pred)}")
```

```python
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Ham', 'Spam']))
 
```

#### Testing on new data

```python
def predict(text):
    vec = tfidf.transform([text])
    result = model.predict(vec)[0]
    return "SPAM" if result == 1 else "HAM"
 
print("\n── Sample Predictions ──")
print(predict("Congratulations! You won a free iPhone. Click now!"))  # SPAM
print(predict("Hey, are we meeting for lunch tomorrow?"))              # HAM
print(predict("URGENT: Claim your $1000 prize today!"))               # SPAM
print(predict("Please review the attached report."))   
print(predict("URGENT: Claim your $1000 prize today!"))
```
